```python
import pandas as pd
import matplotlib.pyplot as plt
HepaC = pd.read_csv('HepatitisCdata.csv')
```

```python
HepaC.head()
```

```python
HepaC.shape
```

```python
HepaC.describe()
```

```python
HepaC['Category'].value_counts()
```

```python
duplicaterows = HepaC[HepaC.duplicated()]
duplicaterows
```

```python
#dropping unnamed column
HepaC.drop(["Unnamed: 0"],axis=1,inplace = True)
```

```python
HepaC.head()
```

```python
HepaC.isnull().sum()
```

```python
#dealing with null entries
HepaC[["ALP"]]=HepaC[["ALP"]].fillna(HepaC[["ALP"]].median())
HepaC[["CHOL"]]=HepaC[["CHOL"]].fillna(HepaC[["CHOL"]].median())
```

```python
HepaC.dropna(subset = ["ALB","ALT","PROT" ], inplace = True)
```

```python
HepaC.isnull().sum()
```

```python
HepaC.shape
```

```python
HepaC['Sex'].value_counts()
```

```python
#replacing categorical data
HepaC['Sex'].replace(['m','f'],[1,0],inplace=True)
HepaC['Sex'].value_counts()
```

```python
replace_dict = dict()
num_dict = dict()
i=0
for val in HepaC['Category'].unique():
    replace_dict[val] = i
    num_dict[str(i)] = val
    i += 1
```

```python
HepaC['Category'].replace(replace_dict,inplace=True)
HepaC.Category.unique()
```

```python
HepaC.head()
```

```python
X = HepaC.drop('Category',axis=1)
y = HepaC['Category']
print("X shape ",X.shape)
print("y shape ",y.shape)
```

```python
#understanding the target column
plt.figure(figsize=(7,7))
HepaC['Category'].value_counts().plot(kind='pie',autopct='%.2f')
plt.title('Category Pie Chart')
plt.show()

HepaC['Category'].value_counts()
```

# Data Smote

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X,y = smote.fit_resample(X,y)
print("X shape ",X.shape)
print("y shape ",y.shape)
```

```python
#Balanced dataset
plt.figure(figsize=(7,7))
y.value_counts().plot(kind='pie',autopct='%.2f')
plt.title('Category Pie Chart')
plt.show()
y.value_counts()
```

```python
numeric_columns = []
for col in HepaC.columns:
    if HepaC[col].dtype == 'int64' or HepaC[col].dtype == 'float64':
        numeric_columns.append(col)
```

```python
#scaling the value
from sklearn.preprocessing import StandardScaler
ssc = StandardScaler()
HepaC[numeric_columns] =ssc.fit_transform(HepaC[numeric_columns])
```

```python
HepaC.head()
```

```python
#spliting dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y ,random_state=25, test_size =0.2)
```

```python
y_train.value_counts(normalize=True)*100
```

```python
y_test.value_counts(normalize =True)*100
```

```python
#stratified sampling
X_train, X_test, y_train, y_test = train_test_split(X, y ,random_state=25, test_size =0.2, stratify = y)
```

# K-nearest neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
knn =KNeighborsClassifier(n_neighbors = 25, )
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
print('Test Accuracy ', accuracy_score(y_test, y_pred ))
print(classification_report(y_test, y_pred))
```

```python
pd.DataFrame(confusion_matrix(y_test, y_pred))
```

```python
print('Train Accuracy ', accuracy_score(y_train, knn.predict(X_train)))
print(classification_report(y_train, knn.predict(X_train)))
```

# Decision tree

```python
from sklearn.tree import DecisionTreeClassifier 
tree_clf = DecisionTreeClassifier(max_depth =5, random_state = 25)
tree_clf.fit(X_train,y_train)
```

```python
y_pred = tree_clf.predict(X_test)
```

```python
print('Test Accuracy ', accuracy_score(y_test, y_pred ))
print(classification_report(y_test,y_pred))
```

```python
print('Train Accuracy ', accuracy_score(y_train, tree_clf.predict(X_train) ))
print(classification_report(y_train, tree_clf.predict(X_train)))
```

```python
from pathlib import Path

IMAGES_PATH = Path() / "images" / "decision_tree"
IMAGES_PATH.mkdir(parents=True, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = IMAGES_PATH / f"{fig_id}.{fig_extension}"
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(
        tree_clf,
        out_file=str(IMAGES_PATH / "hepatitisC.dot"), 
        feature_names=X_train.columns,
        class_names=['0','1','2','3','4'],
        rounded=True,
        filled=True
    )
```

```python
from IPython import display
```

```python
import matplotlib.pyplot as plt
train_acc = []
test_acc = []
max_depth =[]

for depth in range(1, 30):
    clf = DecisionTreeClassifier(max_depth = depth)
    clf.fit(X_train, y_train)
    train_acc.append(accuracy_score(y_train, clf.predict(X_train)).round(3))
    test_acc.append(accuracy_score(y_test, clf.predict(X_test)).round(3))
    max_depth.append(depth)

for i in range(1, 30):
    print('Max Depth = ', i, ', Train Acc. = ', train_acc[i-1], ', Test Acc. = ' ,
         test_acc[i-1])

plt.figure(figsize = (8,6))
plt.xlabel('Max-Depth')
plt.ylabel('Accuracy')
plt.title('Variation of Accuracy with Max-Depth')
plt.plot(max_depth, train_acc, color = 'blue', linewidth = 1,
         label = 'train accuracy');

plt.plot(max_depth, test_acc, color = 'red', linewidth = 2,
         label = 'test accuracy')
plt.legend()
save_fig('Accuracy against max_depth')
plt.show();
```

# Voting classifier (hard voting)

```python
from sklearn.ensemble import  VotingClassifier
voting_clf = VotingClassifier(
    estimators=[
        ('knear', knn),
        ('lr', log_reg),
        ('dt', tree_clf),
        ('svc', Svec)
    ], 
    voting = 'hard'
)
voting_clf.fit(X_train, y_train)
```

```python
for clf in [knn, log_reg, tree_clf, Svec, voting_clf]:
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

# ADA boosting

```python
from sklearn.ensemble import AdaBoostClassifier

ada_clf = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=4), n_estimators=10,
    learning_rate=0.5, random_state=42)
ada_clf.fit(X_train, y_train)
y_pred = ada_clf.predict(X_test)
accuracy_score(y_test, y_pred)
```
